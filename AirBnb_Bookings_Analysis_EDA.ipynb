AirBnb_Bookings_Analysis_EDA

Since 2008, guests and hosts have used Airbnb to expand on travelling possibilities and present a more unique, personalised way of experiencing the world. Today, Airbnb became one of a kind service that is used and recognized by the whole world. Data analysis on millions of listings provided through Airbnb is a crucial factor for the company. These millions of listings generate a lot of data - data that can be analysed and used for security, business decisions, understanding of customers' and providers' (hosts) behaviour and performance on the platform, guiding marketing initiatives, implementation of innovative additional services and much more. This dataset has around 49,000 observations in it with 16 columns and it is a mix of categorical and numeric values. Explore and analyse the data to discover key understandings.


```python
import pandas as pd
import numpy as np
```

```python
file_path='/content/Copy of Airbnb NYC 2019.csv'
df=pd.read_csv(file_path)
df
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df[df['name'].isnull()]
```

```python
df['name'] = df['name'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df[df['host_name'].isnull()]
```

```python
df['host_name'] = df['host_name'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df['last_review'] = df['last_review'].fillna('No Review')
```

```python
df.isnull().sum()
```

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(0.0)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['room_type'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
plt.title("Room Type Share")
plt.show()
```

```python
sns.barplot(data=df, x='neighbourhood_group', y='price', estimator=np.mean)
plt.title("Average Price by Neighbourhood Group")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df, x='neighbourhood_group')
plt.title("Total Listings per Borough")
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=df, x='room_type', y='price', estimator=np.mean)
plt.title("Average Price by Room Type")
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=df[df['price'] < 1000], x='room_type', y='price')
plt.title("Price Distribution (<$1000) by Room Type")
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=df[df['price'] < 500], x='availability_365', y='price', hue='room_type', alpha=0.5)
plt.title("Availability vs Price (<$500)")
plt.tight_layout()
plt.show()
```

```python
sns.histplot(df['minimum_nights'], bins=50)
plt.title("Distribution of Minimum Nights")
plt.xlim(0, 30)
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df, x='neighbourhood_group', hue='room_type')
plt.title("Room Type Distribution by Borough")
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=df[df['price'] < 500], x='number_of_reviews', y='price', hue='neighbourhood_group', alpha=0.6)
plt.title("Price vs Reviews (<$500)")
plt.tight_layout()
plt.show()
```

```python
top_neighbourhoods = df['neighbourhood'].value_counts().nlargest(10).index
sns.countplot(data=df[df['neighbourhood'].isin(top_neighbourhoods)], y='neighbourhood', order=top_neighbourhoods)
plt.title("Top 10 Neighbourhoods with Most Listings")
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=df[df['price'] < 500], x='reviews_per_month', y='price', hue='room_type', alpha=0.6)
plt.title("Price vs Reviews/Month (<$500)")
plt.tight_layout()
plt.show()
```

Conclusion:
Based on the Airbnb NYC 2019 dataset analysis with 49,000+ listings and the visualizations we generated, here are the key conclusions,
Manhattan is the Most Expensive Borough,
Average prices are significantly higher in Manhattan compared to other boroughs like Brooklyn or Queens.
Users looking for budget stays may prefer Brooklyn or Queens.
Private Rooms Are the Most Common Room Type,
Private rooms dominate across boroughs.
Entire homes are common in Manhattan but more expensive.
Users seeking affordability with privacy should look at private rooms in Brooklyn or Queens.
Listings in Brooklyn and Manhattan Dominate the Platform,
These two boroughs have the highest number of listings.
More choices available, especially in central and tourist-friendly areas.
Room Type Influences Availability,
Entire homes/apartments have higher variability in availability (some are always open, others rarely).
Shared rooms are less available and less common.
Most Hosts Set a Minimum Stay of Less Than 7 Days,
Majority of listings have a minimum stay of 1–3 nights, suitable for short-term travelers.
Price Does Not Correlate Strongly With Reviews or Availability,
Higher price ≠ more reviews or better availability.
Some lower-priced listings are highly reviewed, indicating better value.
Neighborhood Popularity Matters,
Certain neighborhoods like Williamsburg, Harlem, and Bedford-Stuyvesant have a high number of listings, indicating popular tourist or residential zones.
Travelers can explore these areas for high availability and diverse pricing.
